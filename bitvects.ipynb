```python
def dists_sq(X, q):
    diffs = X - q
    return np.sum(diffs * diffs, axis=(len(X.shape) - 1)) # handle 1D X

def dists_to_vects(X, q):
    row_norms = np.sum(X*X, axis=1, keepdims=True)
    q_norms = np.sum(q*q, axis=1)
    prods = np.dot(X, q.T)
    return -2*prods + row_norms + q_norms

def randwalk(*args):
    ret = np.random.randn(*args)
    ret = np.cumsum(ret, axis=-1)
    return ret / np.linalg.norm(ret, axis=-1, keepdims=True) * ret.shape[-1]

def top_k_idxs(elements, k, smaller_better=False):
    if smaller_better:
        which_nn = np.arange(k)
        return np.argpartition(elements, kth=which_nn)[:k]
    else:
        which_nn = len(elements) - 1 - np.arange(k)
        return np.argpartition(elements, kth=which_nn)[-k:][::-1]

def hamming_dist(v1, v2):
    return np.count_nonzero(v1 != v2)

def hamming_dists(X, q):
    return np.array([hamming_dist(row, q) for row in X])
```

```python
# okay, so how well does neighbor hash do?

from sklearn import cluster
import kmc2

def kmeans(X, k):
    seeds = seeding = kmc2.kmc2(X, k)
#     plt.imshow(centroids, interpolation=None)
    estimator = cluster.MiniBatchKMeans(k, init=seeding, max_iter=16).fit(X)
#     estimator = cluster.KMeans(k, max_iter=4).fit(X)
    return estimator.cluster_centers_, estimator.labels_

N = int(1e5)
D = 128
Q = 200
num_centroids = 256
num_bits = 64

# data_func = np.random.randn
# data_func = np.random.rand
data_func = randwalk
X = data_func(N, D)
queries = data_func(Q, D)

centroids, labels = kmeans(X, num_centroids)
# plt.imshow(centroids, interpolation=None); print
```

```python
# form groups associated with each centroid
groups = [[] for _ in range(num_centroids)]
for i, lbl in enumerate(labels):
    groups[lbl].append(X[i])

for i, g in enumerate(groups[:]):
    groups[i] = np.array(g)

group_sizes = [len(g) for g in groups]
print group_sizes # huh; these are like 80% singleton clusters in 64D and 32 kmeans iters...
# print sum(group_sizes)
# plt.hist(labels)
plt.hist(group_sizes, bins=num_centroids)
plt.show()

####
# SELF: ensure that there's no group of 0 (which means the kmeans failed) or you'll get 
# mysterious errors later, probably complaining about nans in the hash func plots
####
```

```python
# compute neighbor graph for all centroids
centroid_k = num_bits
centroids_knn_idxs = np.empty((num_centroids, centroid_k), dtype=np.int)
centroids_knn_dists = np.empty((num_centroids, centroid_k))
for i, c in enumerate(centroids):
    dists = dists_sq(centroids, c)
    knn_idxs = top_k_idxs(dists, centroid_k + 1, smaller_better=True)[1:]
    centroids_knn_idxs[i, :] = knn_idxs
    centroids_knn_dists[i, :] = dists[knn_idxs]
```

```python
# compute dists to pivots within each group, and encode
# vects based on dists to the pivots
pivot_dists = [[] for _ in groups]
for i, g in enumerate(groups):
#     print centroids_knn_idxs[i].shape
    assert len(g.shape) == 2
    assert g.shape[1] == D
    pivots = centroids[centroids_knn_idxs[i]]
    pivot_dists[i] = dists_to_vects(g, pivots)
```

```python
thresholds = [[] for _ in groups]
bit_mats = [[] for _ in groups]
for i, g in enumerate(groups):
    thresholds[i] = np.median(pivot_dists[i], axis=0)
    bit_mats[i] = pivot_dists[i] > thresholds[i]
```

```python
def quantize(A, bits=8, max_sigma=-1, normalize=True):
    nbins = int(2 ** bits)
    if max_sigma <=0:
        max_sigma = -stats.norm.ppf(1. / nbins) * (nbins / 2) / ((nbins / 2) - 1) # end bins have 1/nbins of the distro (better)
    A_z = (A - means) / std if normalize else A
    max_val = 2 ** int(bits) - 1
    min_val = -(max_val + 1)
    scale_by = max_val / float(max_sigma)
    quantized = np.floor(A_z * scale_by)
    return np.clip(quantized, min_val, max_val).astype(np.int)

def dbq(A, lower_threshs, upper_threshs):
    # we take sqrt so dist_sq() will yield hamming dist
    return np.sqrt((A > lower_threshs).astype(np.float) + (A > upper_threshs))
#     return (A > lower_threshs).astype(np.float) + (A > upper_threshs)

def fit_dbq_thresholds(A):
#     return np.percentile(A, q=[33, 67], axis=0, interpolation='midpoint')
    return np.percentile(A, q=[33, 67], interpolation='midpoint')

# print A 
# fit_dbq_thresholds(A) # yep, seems to be working
```
